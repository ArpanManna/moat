```python
import argparse
import numpy as np
import pandas as pd
import pickle
import copy
import torch
import torchvision
import matplotlib.pyplot as plt
from time import time
from torchvision import datasets,transforms
from torch import optim
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset, TensorDataset
import os
import random
from tqdm import tqdm
import torch.nn.functional as F
from collections import Counter
from itertools import islice
import itertools
```

```python
with open('comm_rounds_fashion_lfa_threshold_18', 'rb') as f:
    rounds = pickle.load(f)
```

```python
marker = itertools.cycle(('o', 'v', '^', 's', '8', '<', '>','p'))
#marker = itertools.cycle(('o', 'v', '^', '>','p'))
from matplotlib.lines import Line2D
```

```python
with open('accuracy_fashion_lfa_threshold_18', 'rb') as f:
    acc = pickle.load(f)
```

```python
with open('asr_lfa_threshold_18', 'rb') as f:
    asr = pickle.load(f)
```

```python
with open('mcr_lfa_threshold_18', 'rb') as f:
    mcr = pickle.load(f)
```

```python
with open('t_mcr_lfa_threshold_18', 'rb') as f:
    t_mcr = pickle.load(f)
```

```python
threshold_list = [1.5,2.0,2.2,2.5,3.0]
```

```python
for i in range(len(rounds)):
    plt.plot(rounds[i],acc[i],marker=next(marker),label='threshold = {}'.format(threshold_list[i]))
plt.xlabel('communication rounds')
plt.ylabel('accuracy')
plt.grid(color='b', ls = '-.', lw = 0.2)
plt.legend()
plt.savefig('grid_threshold_lfa_acc',dpi=300,bbox_inches='tight')
```

```python
for i in range(len(rounds)):
    plt.plot(rounds[i],asr[i],marker=next(marker),label='threshold = {}'.format(threshold_list[i]))
plt.xlabel('communication rounds')
plt.ylabel('attack success rates')
plt.grid(color='b', ls = '-.', lw = 0.2)
plt.legend()
plt.savefig('grid_threshold_lfa_asr',dpi=300,bbox_inches='tight')
```

```python
for i in range(len(rounds)):
    plt.plot(rounds[i],t_mcr[i],marker=next(marker),label='threshold = {}'.format(threshold_list[i]))

plt.xlabel('communication rounds')
plt.ylabel('targeted misclassification rates')    
plt.grid(color='b', ls = '-.', lw = 0.2)
plt.legend()
plt.savefig('grid_threshold_lfa_tmcr',dpi=300,bbox_inches='tight')
```
